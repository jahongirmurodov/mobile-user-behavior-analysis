## **Анализ пользовательской активности и поведения в мобильных приложениях**

### **1. Цель проекта**

Провести комплексный анализ пользовательских данных мобильного приложения, определить ключевые факторы поведения пользователей, выявить закономерности использования устройств и приложений, а также сформировать рекомендации по улучшению UX и увеличению удержания пользователей.

---

# **2. Описание датасета**

Датасет содержит информацию о поведении пользователей мобильных устройств.

### **Столбцы:**

* **User ID** — уникальный идентификатор пользователя
* **Device Model** — модель устройства
* **Operating System** — Android / iOS
* **App Usage Time (min/day)** — время использования приложений в минутах в день
* **Screen On Time (hours/day)** — время включенного экрана
* **Battery Drain (mAh/day)** — расход батареи
* **Number of Apps Installed** — количество установленных приложений
* **Data Usage (MB/day)** — мобильный трафик
* **Age** — возраст пользователя
* **Gender** — Male / Female / Other
* **User Behavior Class** — классификация поведения (1–4)
